```python
# Polynomial Regression is regression algorithm that models the relationship betweeen dependent(y) and 
# independent variable x as 
# nth degree polynomial
# y=b0+b1x1+.......
# linear model with some modification to increase the accuracy
# used dataset in polynomial regression is of non=linear nature
# used linear regression model
```

```python
# Linear model yields good result on linear dataset but it fails in non-linear dataset so the usage of polynomial came 
# # in case of non-linear dataset
```

```python
# Steps for Polynomial Regrssion
# 1.Data Pre-processing
# 2.Build a linear Regression model and fit it to the dataset
# 3.Build a polynomial regression model and fit to the dataset
# 4.Visualizing the models
# 5.Predicting the output
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
data_set=pd.read_csv('./data_for_polynomial.csv')
data_set
```

```python
#Extracting independent variables(x) and dependent main variable(y)
x=data_set.iloc[:,1:2].values
x
```

```python
y=data_set.iloc[:,2].values
y
```

```python
from sklearn.linear_model import LinearRegression
# Fitting ie. trainint the data with respect to the obj of linearRegression class
lin_regs=LinearRegression()
lin_regs.fit(x, y)
```

```python
# Fitting the Polynomial regression to the dataset
from sklearn.preprocessing import PolynomialFeatures
poly_regs=PolynomialFeatures(degree=2)
# Through the above line, first We are converting our feature matrix into polynomial feature matrix and then fitting it to 
# the Polynomial Regression Model
x_poly=poly_regs.fit_transform(x)
lin_reg_2=LinearRegression()
lin_reg_2.fit(x_poly, y)
```

```python
# Now visualizing the result for LinearRegression model as we did in Simple Linear Regression using the matplotlib library
fig = plt.figure()
fig.patch.set_facecolor('green')
fig.patch.set_alpha(.4)
# Above 2 determines the border background color & it's opacity

ax = fig.add_subplot(111)
ax.plot(range(10))

ax.patch.set_facecolor('black')
ax.patch.set_alpha(1)
# Above 2 determines the inside background color & it's opacity

# If we don't specify the edgecolor and facecolor for the figure when
# saving with savefig, it will override the value we set earlier!
fig.savefig('temp.png', facecolor=fig.get_facecolor(), edgecolor='none')

plt.scatter(x, y, color='blue')
plt.plot(x, lin_regs.predict(x), color='red')
plt.title('False Detection Model(Linear Regresion)')
plt.xlabel('Position Levels')
plt.ylabel('Salary')
plt.show()
```

```python
plt.rcParams.update({
    "figure.facecolor":  (0.0, 1.0, 0.0, 0.2),  # red   with alpha = 30%
    "axes.facecolor":    (0.0, 0.0, 0.0, 1),  # green with alpha = 50%
    "savefig.facecolor": (0.0, 0.0, 0.0, 1),  # blue  with alpha = 20%
})

plt.plot(range(10))
plt.savefig("temp.png")

plt.scatter(x, y, color='blue')
plt.plot(x, lin_reg_2.predict(poly_regs.fit_transform(x)), color='red')
plt.title('Bluff Detection Model(Polynomial Regression)')
plt.xlabel('Positions')
plt.ylabel('Salary')
plt.show()
```

```python
# Fitting the Polynomial regression to the dataset
from sklearn.preprocessing import PolynomialFeatures
poly_regs=PolynomialFeatures(degree=4)
# Through the above line, first We are converting our feature matrix into polynomial feature matrix and then fitting it to 
# the Polynomial Regression Model
x_poly=poly_regs.fit_transform(x)
lin_reg_2=LinearRegression()
lin_reg_2.fit(x_poly, y)
```

```python
plt.rcParams.update({
    "figure.facecolor":  (0.0, 1.0, 0.0, 0.2),  # red   with alpha = 30%
    "axes.facecolor":    (0.0, 0.0, 0.0, 1),  # green with alpha = 50%
    "savefig.facecolor": (0.0, 0.0, 0.0, 1),  # blue  with alpha = 20%
})

plt.plot(range(10))
plt.savefig("temp.png")

plt.scatter(x, y, color='blue')
plt.plot(x, lin_reg_2.predict(poly_regs.fit_transform(x)), color='red')
plt.title('Bluff Detection Model(Polynomial Regression) with Degree 4')
plt.xlabel('Positions')
plt.ylabel('Salary')
plt.show()
```

```python
# Predicting final result with linear regression model
lin_pred = lin_regs.predict([[6.5]])
print(lin_pred)
```

```python
# Predicting final result with Polynomial Regression Model
poly_pred = lin_reg_2.predict(poly_regs.fit_transform([[6.5]]))
print(poly_pred)
# the output in polynomial regression is much closer to original value so we can say polynomial is better that linear
```
