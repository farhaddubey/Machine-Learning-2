```python
import mglearn
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
mpl.rcParams.update({
    "figure.facecolor":"black",
    "axes.facecolor":"black",
    "axes.edgecolor":"orange",
    "xtick.color":"red",
    "ytick.color":"yellow",
    "legend.labelcolor":"linecolor"
})
plt.plot(range(10))
```

```python
mglearn.plots.plot_linear_regression_wave()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X, y=mglearn.datasets.make_wave(n_samples=60)
X_train, X_test, y_train, y_test=train_test_split(X, y, random_state=0)
linreg=LinearRegression()
linreg.fit(X_train, y_train)
```

```python
print(linreg.coef_)
# coefficent attribute is numpy array
```

```python
print(linreg.intercept_)
# while intercept attribute is floating no.
```

```python
# Now let's look at the performance of the training and test set
print("Training set score: ", linreg.score(X_train, y_train))
print("Testing set score: ", linreg.score(X_test, y_test))
```

```python
# As the score of R**2 is .69 which is not very good indication fo model and the scores of both training & test set are close
# so We are likely to underfit. Examples of Underfittin that's.
```

```python
# Now testing linear regression model using Mglearn's Boston dataset
X, y=mglearn.datasets.load_extended_boston()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
linreg=LinearRegression()
linreg.fit(X_train, y_train)
```

```python
linreg.score(X_train, y_train)
```

```python
linreg.score(X_test, y_test)
```

```python
# Here thers's a significant discrepancy between, training set and the test set. which is a clear sign of Overfitting.
# Ridge Regression is standard alternatives to LinearRegression
```
