```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.rcParams.update({
    "figure.facecolor":"black",
    "axes.facecolor":"black",
    "axes.edgecolor":"green",
    "xtick.color":"yellow",
    "ytick.color":"red",
    "legend.labelcolor":"linecolor",
})
```

```python
import mglearn
from sklearn.datasets import make_blobs
X, y=make_blobs(random_state=42)
mglearn.discrete_scatter(X[:,0], X[:,1], y)
plt.xlabel("Feature 0", c='y')
plt.ylabel("Feature 1", c='r')
plt.legend(["Class 0", "Class 1", "Class 2"])
```

2D toy dataset containing three classes

```python
import warnings
warnings.filterwarnings('ignore')
#Next Onwards no warning will be shown
```

```python
from sklearn.svm import SVC
from sklearn.svm import LinearSVC
linear_svm=LinearSVC().fit(X, y)
print("Coefficient Shape:", linear_svm.coef_.shape)
print("Intercept shape:", linear_svm.intercept_.shape)
```

```python
mglearn.discrete_scatter(X[:,0], X[:,1], y)
# discrete_scatter is used to create variable where X is the input and y the target variable
# represents feature value along 1st(0) & 2nd(1) axis; y represents target variable in which class/category each data
# belongs to
line=np.linspace(-15, 25)
# creates a evenly space lines, used to represt the decision boundary
for coef, intercept, color in zip(linear_svm.coef_, linear_svm.intercept_, ['b', 'r', 'g']):
    plt.plot(line, -(line*coef[0]+intercept)/coef[1], c=color)
plt.ylim(-10, 15)
plt.xlim(-10, 8)
plt.xlabel("Feature 0", c='b')
plt.ylabel("Feature 1", c='r')
plt.legend(["Class 0", "Class 1", "Class 2", "Line Class 0", "Line Class 1", "Line Class 2"], loc=(1.01, .3))
```

```python
mglearn.plots.plot_2d_classification(linear_svm, X, fill=True, alpha=.7)
mglearn.discrete_scatter(X[:,0], X[:,1], y)
line=np.linspace(-15, 15)
for coef, intercept, color in zip(linear_svm.coef_, linear_svm.intercept_, ['b', 'r', 'g']):
    plt.plot(line, -(line*coef[0] + intercept)/coef[1], c=color)
plt.legend(["Class 0", "Class 1", "Class 2", "Line class 0", "Line Class 1", "Line Class 2"], loc=(1.01, .3))
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

```python
mglearn.plots.plot_2d_classification(linear_svm, X, fill=False, alpha=.4)
mglearn.discrete_scatter(X[:,0], X[:,1], y)
line=np.linspace(-15, 15)
for coef, intercept, color in zip(linear_svm.coef_, linear_svm.intercept_, ['b', 'r', 'g']):
    plt.plot(line, -(line*coef[0] + intercept)/coef[1], c=color)
plt.legend(["Class 0", "Class 1", "Class 2", "Line class 0", "Line Class 1", "Line Class 2"], loc=(1.01, .3))
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```
