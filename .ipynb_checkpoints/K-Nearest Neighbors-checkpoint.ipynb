```python
from sklearn.model_selection import train_test_split
X, y=mglearn.datasets.make_forge()
X_train, X_test, y_train, y_test=train_test_split(X, y, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knnClassifier=KNeighborsClassifier(n_neighbors=3)
knnClassifier.fit(X_train, y_train)
```

```python
knnClassifier.predict(X_test)
```

```python
knnClassifier.score(X_test, y_test)
```

```python
fig, axes=plt.subplots(1, 4, figsize=(15, 4))
plt.rcParams.update({
    'figure.facecolor':'black',
    'axes.edgecolor':'orange', 
    'xtick.color':'white', 
    'ytick.color':'green', 
    'legend.labelcolor': 'linecolor'    
})
mlp.rc('axes',edgecolor='green')
for n_neighbors, ax in zip([1, 3, 9, 15], axes):
    knnClassifier=KNeighborsClassifier(n_neighbors=n_neighbors).fit(X, y)
    mglearn.plots.plot_2d_separator(knnClassifier, X, fill=True, eps=0.5, ax=ax, alpha=.4)
    mglearn.discrete_scatter(X[:,0],X[:, 1], y, ax=ax)
    ax.set_title("{} neighbors".format(n_neighbors))
    ax.set_xlabel("feature 0")
    ax.set_ylabel("feature 1")
axes[0].legend(loc=3)
```

```python
plt.rcParams.update({
    'figure.facecolor':'black',
    'axes.edgecolor':'orange', 
    'xtick.color':'green', 
    'ytick.color':'green', 
    'legend.labelcolor': 'linecolor'    
})
mlp.rc('axes',edgecolor='green')
from sklearn.datasets import load_breast_cancer
cancer_dataset=load_breast_cancer()
X_train, X_test, y_train, y_test=train_test_split(cancer_dataset.data, cancer_dataset.target, 
                                                  stratify=cancer_dataset.target, random_state=0)
training_accuracy=[]
test_accuracy=[]
# Trying n neighbors from 1 to 10
neighbors_list=range(1, 11)
for n_neighbors in neighbors_list:
    # Build the model
    knnClassifier=KNeighborsClassifier(n_neighbors=n_neighbors)
    knnClassifier.fit(X_train, y_train)
    # Record training set Accuracy
    training_accuracy.append(knnClassifier.score(X_train, y_train))
    # Record Generalisation Accuracy
    test_accuracy.append(knnClassifier.score(X_test, y_test))
plt.plot(neighbors_list, training_accuracy, label='Training Accuracy')
plt.plot(neighbors_list, test_accuracy, label='Test Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('n_neighbors')
plt.legend()
    
```

```python
# Scikit Learn's KNeighborsRegressor is similar like KNeighborsClassifier 
from sklearn.neighbors import KNeighborsRegressor
X, y=mglearn.datasets.make_wave(n_samples=40)
X_train, X_test, y_train, y_test=train_test_split(X, y, random_state=0)
knnRegressor=KNeighborsRegressor(n_neighbors=3)
knnRegressor.fit(X_train, y_train)
```

```python
# DAta is fitted ie. trained Now we can make the predictions
print(knnRegressor.predict(X_test))
```

```python
print(knnRegressor.score(X_test, y_test))
```

```python
fig, axes=plt.subplots(1, 3, figsize=(15, 4))
import numpy as np
line = np.linspace(-3, 3, 1000).reshape(-1, 1)
for n_neighbors, ax in zip([1, 3, 9], axes):
    #make predictions using 1, 3, 9 neighbors
    reg=KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(X_train, y_train)
    ax.plot(line, reg.predict(line))
    ax.plot(X_train, y_train, '^', c=mglearn.cm2(0), markersize=8)
    ax.plot(X_test, y_test, '*', c=mglearn.cm2(1), markersize=8)
    
    ax.set_title(
    "{} neighbor(s)\n train score: {:.2f} test score: {:.2f}".format(n_neighbors, reg.score(X_train, y_train),
                                                                     reg.score(X_test, y_test)))
    plt.xlabel("Feature")
    plt.ylabel("Target")
    axes[0].legend(["Model predictions", "Training Data/Target", "Test Data/target"], loc="best")
```
