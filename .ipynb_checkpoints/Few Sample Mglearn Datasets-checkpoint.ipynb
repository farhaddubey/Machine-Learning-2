```python
import mglearn
# forge dataset is synthetic two class dataset
X, y=mglearn.datasets.make_forge()
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
import matplotlib.pyplot as plt
import matplotlib as mlp
plt.rcParams.update({
    "figure.facecolor":(0, 1, 0, 0.2),
    "axes.facecolor":(0, 0, 0, 1),
})
mlp.rc('axes',edgecolor='green')
plt.plot(range(10))
plt.legend(["Class 0", "Class 1"], loc=4)
plt.xlabel("1st Features")
plt.ylabel("2nd Features")
print(X.shape)
print(X)
```

```python
X, y=mglearn.datasets.make_wave(n_samples=40)
plt.plot(X, y, 'o')
plt.ylim(-3, 3)
plt.xlabel("Feature")
plt.ylabel("Target")
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
print(cancer.keys())
```

```python
print(cancer.data.shape)
```

```python
print("Sample counts per class:\n {}".format(cancer.data.shape))
```

```python
import numpy as np
print("Sample counts per class: \n{}".format({n:v for n, v in zip(cancer.target_names, np.bincount(cancer.target))}))
```

```python
print("Feature Names: {}".format(cancer.feature_names))
```
