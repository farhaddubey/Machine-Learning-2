<h2 style="color:red">K-Means Clustering Algorithm</h2>
<p style="color:blue">an Unsupervised Learning Algorithm, which groups the unlabeled dataset into different clusters.
Here, k defines the no.. of clusters required to be created in the process</p>

<p style="color:blue">an iterative algorithm, that divides the unlabeled dataset into k different clusters in 
such a way that each dataset
belongs only one group that has similar properties</p>


<p style="color:teal">allows us to cluster the data into different groups and a convenient way to discover the categories
of groups in the unlabeled dataset on its own without the need of any training</br>
reduces distance between the data point and their cluster</p>
<p style="color:yellow">k-means algo performs two main tasks</p>
<li style="color:blue">Determining the best value for k center points or centroieds by an iterative process</li>
<li style="color:orange">Assigning each data point to its closest k=center</li>

```python
# Python Implementation of K -Means Algorithm
# 1. Data Preprocessing
# 2. Finding the optimal no. of clusters using the elbow method
# 3. Training the k-means algorithm on the training dataset
# 4. Visualizing the clusters
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update({
    "figure.facecolor":"black",
    "figure.facecolor":"black",
    "figure.edgecolor":"orange",
    "axes.facecolor":"black",
    "axes.edgecolor":"green",
    "xtick.color":"red",
    "ytick.color":"yellow",
    "legend.labelcolor":"linecolor"
})
```

```python
dataset=pd.read_csv("A:/Machine Learning/Mall_Customers.csv")
pd.set_option('display.max_rows', None)
dataset
```

```python
# Extracting independent variables
x=dataset.iloc[:,[3, 4]].values
x
```

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.cluster import KMeans
wcss_list = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(x)
    wcss_list.append(kmeans.inertia_)
plt.plot(range(1,11), wcss_list)
plt.title('The Elbow Method Graph', c='blue')
plt.xlabel('Number of Clusters', c='red')
plt.ylabel('WCSS_list', c='yellow')
plt.show()
```

```python
wcss_list
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_predict=kmeans.fit_predict(x)
y_predict
```

```python
plt.scatter(x[y_predict==0,0], x[y_predict==0, 1], s=100, c='blue', label='Cluster 1')
plt.scatter(x[y_predict==1,0], x[y_predict==1, 1], s=100, c='green', label='Cluster 2')
plt.scatter(x[y_predict==2,0], x[y_predict==2, 1], s=100, c='red', label='Cluster 3')
plt.scatter(x[y_predict==3,0], x[y_predict==3, 1], s=100, c='cyan', label='Cluster 4')
plt.scatter(x[y_predict==4,0], x[y_predict==4, 1], s=100, c='magenta', label='Cluster 5')
plt.title('Cluster for customers', c='g')
plt.xlabel('Annual Income(k$)', c='r')
plt.ylabel('Spending Score (1-100)', c='y')
plt.legend()
plt.show()
```
